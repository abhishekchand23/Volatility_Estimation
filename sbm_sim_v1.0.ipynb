```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
def prachi_estimator(So,mu,sigma,N,ins):
    data_fin=pd.DataFrame()
    for i in range(ins):
        b=mu*(1/N)+sigma*np.sqrt(1/N)*np.random.normal(0,1,N)
        b=np.insert(b,0,So,axis=None)
        zz=np.cumsum(b)
        temp=pd.DataFrame({'Open':zz[0],
             'Close':zz[-1],
             'High':max(zz),
             'Low':min(zz)},index=pd.MultiIndex.from_tuples([('Simulation'+" "+ str(i+1),"mu={}, sigma_sq={}".format(mu,round(sigma**2,2)))]))
        data_fin=pd.concat([data_fin,temp])
        print("loop {} done at {}".format(i+1,datetime.datetime.now().time()))
    return data_fin
```

```python
So=0
mu=0
var=0.2
sigma=np.sqrt(var)
N=700000
ins=20000
final=prachi_estimator(So,mu,sigma,N,ins)
```
